```python
# Load the dataset
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Startups_Data.csv')
data.head()
# need to convert object into numerical because we use regression
# can't delete it because it has relationship with the cost
```

```python
data.info()
```

```python
# Data Preparation
# select dependent and independent variable

data.columns
```

```python
x = ['R&D Spend', 'Administration', 'Marketing Spend', 'State']
x = data[x].values
x
# below is independent variables
```

```python
y = data.Profit.values
y
# profit is dependent variable
```

```python
#OneHotEncoding

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

transformer = ColumnTransformer(
    transformers = [("OneHot",OneHotEncoder(),[3])],
    remainder ='passthrough'  # please do not remove other columns which is not specified.refer columntransformer scikitlearn

)

x = transformer.fit_transform(x)
x = x.astype('int')
```

```python
x
# 001 --> new york
# 100 --> california
# 010 --> florida
# 0 0 1 r&d admin marketing
```

```python
# https://pandas.pydata.org/docs/reference/api/pandas.get_dummies.html
# pandas_dummies
```

```python
# Split the dataset into 2 parts
# a.) Training set  : apply a ML algo on training set
# b.) Testing set   : we will be using for testing / Evaluation

# 0.2 --> 20% of the data will be used for testing purposes
# and remaining 80% of the data will be used for training
# depends on your project requirement on how many % you want to do training and testing
# the more data you train, the better result you get

# random_state --> random shuffling
# random_state=42 --> locked the randomness.
# without 42, the number will be changed everytime we refresh and run all the cell again

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print('Original Shape')
print(x.shape)
print(y.shape)
print("\nAfter Train Test Split")  # \n --> for the space above the code
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)

# 6 --> the nummber of columns
# 50 by 6 matrix
```

```python
# Model Development
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(x_train,y_train)
```

```python
# pickle can be used to save our trained model
import pickle
pickle.dump(mlr,open('multiple_reg.pkl','wb'))  # w->write   b->byte stream
# 'multiple_reg.pkl --> it's just file name. can give any name to it. but need to include .pkl
```

```python
# Model Evaluation
pred_values = mlr.predict(x_test)
print(pred_values)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,pred_values)
print("Score Of The Model : ",score)

# r2 tells you how close your predicted numbers are to the real numbers overall
# r2 is a regression performance metrics
# r2 coefficient of determination
# r2 is a number between 0 and 1

# > 0.9        Excellent Fit
# 0.7 - 0.9    Good Fit
# 0.5 - 0.7    Moderate Fit
# <0.5         Weak Fit
```
